## Cleaning Data

```python
df_ecommerce.info()
```

```python
#karena terdapat description yang kosong, customerid juga kosong. maka diperlukan manipulasi data
#description menjadi penting, dikarenakan jika kosong kita bisa menduga bahwa itu tidak ada nama barangnya (tidak mungkin kosong)
df_ecommerce[df_ecommerce['Description'].isnull()]
```

```python
df_ecommerce = df_ecommerce.dropna(subset='Description')
df_ecommerce
```

```python
#kita juga coba checking customerid. customer id haruslah ada, karena jika tidak ada berarti tidak ada data pembeli (tidak mungkin)
df_ecommerce[df_ecommerce['CustomerID'].isnull()]
```

```python
# kita tidak dapat menghapus customerid yang kosong, dikarenakan data yang cukup banyak (133626 rows). diganti saja dengan unkonwn
df_ecommerce['CustomerID'] = df_ecommerce['CustomerID'].fillna(00000)
df_ecommerce[df_ecommerce['CustomerID'].isnull()]
```

```python
df_ecommerce
```

```python
df_ecommerce.info()
```

```python
#karena invoicedate bertipe data object, maka kita perlu mengubahnya menjadi datetime
df_ecommerce['InvoiceDate']= df_ecommerce['InvoiceDate'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
df_ecommerce.info()
```

```python
df_ecommerce
```

```python
#cek duplikasi data
df_ecommerce.loc[df_ecommerce.duplicated()]
```

```python
df_ecommerce = df_ecommerce.drop_duplicates()
```

```python
df_ecommerce
```

```python
#untuk memudahkan nanti dalam analisa, kita bisa melakukan lowercse pada data Description dan Country
df_ecommerce['Description'] = df_ecommerce['Description'].apply(lambda x: x.lower())
df_ecommerce['Country'] = df_ecommerce['Country'].apply(lambda x: x.lower())
```

```python
df_ecommerce
```

```python
#ubah tipe data customerid menjadi integer
df_ecommerce['CustomerID'] = df_ecommerce['CustomerID'].astype(int)
```
